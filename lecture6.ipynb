```python
import nltk
nltk.download('book')
```

```python
from nltk.book import *
text4.concordance("freedom")
```

```python
#similar(): Finds words that appear in similar contexts to a given word.
text4.similar("freedom")

```

```python
#common_contexts(): Shows the shared contexts in which two or more words appear. Examples
text4.common_contexts(["freedom", "liberty"])
```

```python
#Dispersion Plot: Create a dispersion plot using a text of your choice with different words. This visualizes the occurrences of words throughout the text. Example:
text4.dispersion_plot(["freedom", "democracy", "justice", "rights"])
```

---
#### Lexical diversity: Lexical diversity measures the ratio of unique words (vocabulary size) to the total number of words in a text. It gives an idea of how rich the vocabulary is within a text. Higher lexical diversity indicates a more varied vocabulary.
---

### Frequency Distributions and Cumulative Frequency Plot: Create frequency distributions using FreqDist and then plot the cumulative frequency of the 75 most common words in text2. Example:

```python
from nltk import FreqDist
fdist2 = FreqDist(text2)
fdist2.plot(75, cumulative=True)
```

---
### Bigram and Collocations: A bigram is a pair of consecutive words in a text. Collocations are word pairs that occur together unusually often. Bigrams can help identify collocations. Example of extracting bigrams:
---

```python
from nltk import bigrams
list(bigrams(text2))
```
