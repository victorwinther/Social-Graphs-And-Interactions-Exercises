```python
import requests

# Function to fetch Wikipedia page source using the API
def fetch_wikipedia_source(page_title):
    url = "https://en.wikipedia.org/w/api.php"
    params = {
        'action': 'query',
        'format': 'json',
        'prop': 'revisions',
        'rvprop': 'content',
        'titles': page_title,
        'formatversion': 2
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    if 'query' in data:
        pages = data['query']['pages']
        if len(pages) > 0 and 'revisions' in pages[0]:
            return pages[0]['revisions'][0]['content']
    return None

# List of my favorite Wikipedia pages
favorite_pages = [
    "Python (programming language)", 
    "Artificial intelligence", 
    "Quantum computing", 
    "Machine learning"
]

# Download sources for each of the favorite pages
sources = {}
for page in favorite_pages:
    print(f"Fetching source for: {page}")
    source = fetch_wikipedia_source(page)
    if source:
        sources[page] = source
        print(f"Successfully fetched source for {page}")
    else:
        print(f"Failed to fetch source for {page}")

# Display the first 500 characters of each page's source
for page, source in sources.items():
    print(f"\nSource for {page}:\n{source[:500]}...\n")
```
